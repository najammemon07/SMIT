```python
# Capitalize Method

name = "pakistan"

print(name.capitalize())
```

```python
# CaseFold Method

text = "straBe"

print(text.casefold())
```

```python
# Center Method

text = "Python"

print(text)

print(text.center(20,"-"))
```

```python
# Count Method

text = "banana"

print(text)
print(text.count("a"))
```

```python
# EndsWith Method

txt = "Python Programming"

print(txt)

# use endswith method

print(txt.endswith("Programming"))
```

```python
# find method

text = "Python Program"

print(text)

# use find method

print(text.find("Program"))
```

```python
# f-string f-format 

name = "Najam"
age = 30

text = f"My name is {name} and I am {age} years old"

print(text)
```

```python
# index method 

txt = "python language"

print(txt.index("l"))
```

```python
# isalnum method

text1 = "Najam1995"
text2 = "Najam@gmail.com"

print(text1.isalnum())
print(text2.isalnum())
```

```python
# isalpha Method checked only string not a number.

text1 = "Python"
text2 = "Python 2.0"

print(text1)

# used isalpha Method

print(text2.isalpha())
```

```python
# ASCII Method me srf letter number or symbols use kr skte hy 
# koi special charactor nhe jse k Urdu or Arabic language Emoji ya special charactors

# this is ASCII method

text1 = "Python 2.0 @"

# this is not ASCII becouse there using emoji 

text2 = "Python😊"

print(text1.isascii())
print(text2.isascii())

```
